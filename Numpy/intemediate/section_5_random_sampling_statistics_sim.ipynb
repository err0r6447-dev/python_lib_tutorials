# ðŸ§  Section 5: Random Sampling and Statistical Simulation

NumPy's `random` module enables reproducible random number generation and large-scale statistical simulation.
We'll explore the modern `numpy.random.default_rng` API for sampling, probability distributions, and vectorized stochastic experiments.

We'll use simulations for realistic cases: data bootstrapping, Monte Carlo approximation, and probabilistic modeling.

## ðŸŽ² 1. The Random Generator API

The older `np.random` API is still functional, but the **new generator** system introduced in NumPy 1.17 offers better reproducibility, isolation, and performance.

**Key idea:** You create a generator instance, which holds its own internal state â€” this allows for deterministic random streams independent from global state.

````python
import numpy as np

# Create a reproducible random number generator (RNG)
rng = np.random.default_rng(seed=42)

# Generate random samples
uniform_samples = rng.random(5)     # Uniformly distributed in [0,1)
int_samples = rng.integers(0, 100, size=5)  # Random integers

print("Uniform samples:", uniform_samples)
print("Integer samples:", int_samples)
````

## ðŸ“Š 2. Sampling from Distributions

NumPy provides many built-in probability distributions for simulation and data modeling: normal, binomial, Poisson, exponential, etc.

You can efficiently generate millions of samples in one call â€” much faster than Python's standard `random` library.

````python
# Normal (Gaussian) distribution
normal_data = rng.normal(loc=0, scale=1, size=10000)

# Binomial distribution: number of successes in n trials
binomial_data = rng.binomial(n=10, p=0.5, size=10000)

# Exponential distribution (time between events)
exp_data = rng.exponential(scale=1.0, size=10000)

print(f"Normal mean/std: {normal_data.mean():.3f}, {normal_data.std():.3f}")
print(f"Binomial mean: {binomial_data.mean():.3f}")
print(f"Exponential mean: {exp_data.mean():.3f}")
````

## ðŸ§© 3. Vectorized Sampling for Simulations

Monte Carlo simulations rely on random sampling to approximate mathematical expectations.
Let's estimate Ï€ by simulating random points inside a square and counting how many fall within a unit circle.

````python
N = 1_000_000
x = rng.random(N)
y = rng.random(N)

# Count points inside unit circle (x^2 + y^2 <= 1)
inside = (x**2 + y**2) <= 1
pi_estimate = 4 * np.sum(inside) / N

print(f"Estimated Ï€ â‰ˆ {pi_estimate:.5f}")
````

## ðŸ§® 4. Bootstrapping and Random Sampling from Data

Random sampling is widely used in data science for resampling methods like **bootstrapping**, where we repeatedly draw samples (with replacement) to estimate uncertainty or variance.

````python
# Example dataset: daily sales values
sales = np.array([210, 180, 230, 260, 250, 275, 300])

# Bootstrap resampling: 1000 random samples with replacement
boot_means = [rng.choice(sales, size=len(sales), replace=True).mean() for _ in range(1000)]

print(f"Bootstrap mean estimate: {np.mean(boot_means):.2f}")
print(f"Bootstrap std estimate: {np.std(boot_means):.2f}")
````

## âš™ï¸ Under the Hood: The PCG64 Bit Generator

- The modern `Generator` uses **PCG64**, a high-quality pseudo-random algorithm.
- It generates 64-bit integers which are transformed into floating-point numbers as needed.
- Independent streams can be created without interfering with global state.
- Vectorized generation avoids Python loops, drastically improving performance.


## âœ… Best Practices & Pitfalls

âœ” **Use `default_rng()` instead of `np.random`** â€” it provides reproducible, isolated streams.
âœ” Always set a **fixed seed** when comparing results or running experiments.
âœ” Prefer **vectorized random draws** (`rng.normal(size=10000)`) instead of looping.
âœ” Use **`rng.choice`** for sampling from existing data.

**Common mistakes:**
- Forgetting to seed â†’ results vary across runs.
- Using `np.random` (legacy API) â†’ less reproducible.
- Confusing `np.random.shuffle` (in-place) with `rng.choice` (returns new array).
